# Data Preprocessing :

      Data preprocessing is a crucial step in the data analysis process that involves preparing raw data into a clean and structured format suitable for analysis and modeling. In this project, several preprocessing techniques are applied to improve data quality, consistency, and usability.

     Initially, column names are cleaned by removing spaces, converting them to lowercase, and eliminating special characters. This ensures uniformity and avoids errors during data handling. Missing values are then addressed using forward fill and backward fill methods, which help in maintaining data continuity without losing important information.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# ---- STEP 1: Clean column names (IMPORTANT) ----
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns = df.columns.str.replace(r"[()]", "", regex=True)

# ---- STEP 2: Handle missing values (FIXED) ----
df = df.ffill()   # instead of fillna(method='ffill')

# Optional: if still nulls exist
df = df.bfill()

# ---- STEP 3: Remove duplicates ----
df = df.drop_duplicates()

# ---- STEP 4: Separate categorical & numeric ----
cat_cols = df.select_dtypes(include=['object']).columns
num_cols = df.select_dtypes(include=['int64','float64']).columns

# ---- STEP 5: Encode categorical columns (SAFE) ----
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# ---- STEP 6: Feature Scaling (IMPORTANT FIX) ----
scaler = StandardScaler()

# Avoid scaling target column (example: salary)
target_col = 'salary' if 'salary' in df.columns else None

scale_cols = df.select_dtypes(include=['int64','float64']).columns

if target_col:
    scale_cols = scale_cols.drop(target_col)

df[scale_cols] = scaler.fit_transform(df[scale_cols])

print("Preprocessing completed successfully ✅")
```

# Visualizations :

     Visualization means representing data in a graphical or visual form so that it is easier to understand and interpret. Instead of looking at raw numbers in tables, visualization uses charts, graphs, and plots to show patterns, trends, and relationships in the data.

    In your project, visualization helps to clearly see how factors like experience, job role, and skills affect salary. For example, a bar chart can show which job role has the highest salary, and a line chart can show how salary increases with experience.

    Visualizations make complex data simple and easy to understand, especially for non-technical users. They also help in identifying patterns, outliers, and important insights quickly. Overall, visualization is an important step in data analysis that helps in better decision-making and communication of results.

Simple Examples:

      ✔ Bar Chart → Compare job roles and salary
      ✔ Line Chart → Show salary trend over experience
      ✔ Pie Chart → Show company size distribution
      ✔ Histogram → Show salary distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns = df.columns.str.replace(r"[()]", "", regex=True)

# Fix data types
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')
df['experience_years'] = pd.to_numeric(df['experience_years'], errors='coerce')

# Drop missing
df = df.dropna(subset=['salary','experience_years'])

# Style
sns.set(style="whitegrid")
```

 💡Salary Trend by Experience :

     ✅ This visualization shows how salary changes as a person gains more experience over time.
     ✅ The data is grouped by years of experience, and the average salary is calculated for each group to get a clear overall trend.
     ✅ The horizontal axis (X-axis) represents the number of years of experience, while the vertical axis (Y-axis) represents the average salary.
     ✅ Each point on the line shows the average salary for a particular experience level, and the line connects these points to show the trend.
     ✅ If the line moves upward, it indicates that salary increases as experience increases.
     ✅ This graph helps people easily understand how earnings grow over time without looking at complex data tables.
     ✅ It is useful for students, job seekers, and professionals to understand career growth and salary expectations.
    

```python
# Salary Trend by Experience (Line Plot)
avg_salary = df.groupby('experience_years')['salary'].mean().reset_index()

plt.figure(figsize=(8,5))
plt.plot(avg_salary['experience_years'], avg_salary['salary'])
plt.title("Salary Trend by Experience")
plt.xlabel("Experience (Years)")
plt.ylabel("Average Salary")
plt.show()
```

💡Salary Distribution (Histogram):

         ✅ This visualization shows how salaries are distributed across all employees in the dataset.
         ✅ It uses a histogram, which groups salaries into ranges and shows how many people fall into each range.
         ✅ The horizontal axis (X-axis) represents different salary ranges (low to high)
         ✅ The vertical axis (Y-axis) shows the number of people (frequency) in each salary range.
         ✅ The bars in the graph help identify which salary ranges have more employees.
         ✅ The smooth curve (KDE line) gives a clearer view of the overall salary pattern and distribution.
         ✅ It helps to easily identify whether most people earn low, medium, or high salaries.
         ✅ This visualization also helps in spotting unusual values or extreme salaries (outliers).
         ✅ Overall, it makes it simple to understand salary patterns without analyzing raw data.
        

```python
#Salary Distribution (Histogram)
plt.figure(figsize=(8,5))
sns.histplot(df['salary'], kde=True)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.show()
```

 💡KDE Plot of Salary Distribution(KDE plot)

         ✅ This visualization shows how salaries are spread across the dataset using a smooth curve instead of bars.
         ✅ It uses a KDE (Kernel Density Estimation) plot, which represents the distribution of salary in a continuous and easy-to-understand way.
         ✅ The horizontal axis (X-axis) represents the salary values from low to high.
         ✅ The vertical axis (Y-axis) represents density, which shows how concentrated the salaries are in certain ranges.
         ✅ The smooth curve helps in understanding the overall pattern of salary distribution more clearly than a histogram.
         ✅ The filled area under the curve makes it easier to visualize where most salaries are concentrated.
         ✅ Peaks in the curve indicate salary ranges where more people are earning similar amounts.
         ✅ It helps identify whether salaries are evenly distributed or concentrated in specific ranges.
         ✅ This visualization is useful for understanding trends and patterns without focusing on individual data points.
        

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['salary'], fill=True)

plt.title("KDE Plot of Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Density")
plt.show()
```

💡Cumulative  salary Growth:

      ✅ Purpose: To visualize how total salary accumulates with increasing experience.
      ✅ Data Sorting: df.sort_values('experience_years') ensures experience is in order for cumulative calculation.
      ✅ Cumulative Sum: cumsum() calculates the running total of salaries.
      ✅ Plotting: plt.plot() draws experience on x-axis and cumulative salary on y-axis.
      ✅ Insights: Helps identify overall salary growth trends and how experience contributes to total earnings.

```python
#Cumulative Salary Growth
sorted_df = df.sort_values('experience_years')

plt.figure(figsize=(8,5))
plt.plot(sorted_df['experience_years'],
         sorted_df['salary'].cumsum())

plt.title("Cumulative Salary Growth")
plt.xlabel("Experience")
plt.ylabel("Cumulative Salary")
plt.show()
```
